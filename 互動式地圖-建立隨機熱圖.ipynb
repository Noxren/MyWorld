```python
import numpy as np
```

```python
data = (np.random.normal(size = (200,2))* np.array([[0.02, 0.02]])+ np.array([[24.987670, 121.577300]]))
```

```python
data = [data.tolist()]
```

```python
for i in range(20):
    data.append((data[i] + np.random.normal(size = (200, 2))*0.001).tolist())
```

```python
from folium.plugins import HeatMapWithTime
```

```python
amap = folium.Map(location = [24.987670, 121.577300], zoom_start = 11)
amap.add_child(HeatMapWithTime(data))
```
