# 挑戰題二:計算GPA

```python
sum = 0
i = 0
one = 1
while(one == 1):
    str1 = input("輸入你的成績(輸入GPA開始計算): ")
    if str1 == "GPA":
        break
    i = i + 1
    S = float(str1)
    if S<=100 and S>=90:
        sum = sum + 4.3
    elif S<=89 and S>=85:
        sum = sum + 4
    elif S<=84 and S>=80:
        sum = sum + 3.7
    elif S<=79 and S>=77:
        sum = sum + 3.3
    elif S<=76 and S>=73:
        sum = sum + 3
    elif S<=72 and S>=70:
        sum = sum + 2.7
    elif S<=69 and S>=67:
        sum = sum + 2.3
    elif S<=66 and S>=63:
        sum = sum + 2
    elif S<=62 and S>=60:
        sum = sum + 1.7
    elif S<=59 and S>=57:
        sum = sum + 1.3
    elif S<=56 and S>=53:
        sum = sum + 1
    elif S<=52 and S>=50:
        sum = sum + 0.7
    elif S<=49 and S>=0:
        sum = sum
GPA = sum / i
print("GPA= " + str(GPA))
```
