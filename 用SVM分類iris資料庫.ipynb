```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import datasets

from sklearn.svm import SVC
```

```python
iris = datasets.load_iris()
```

```python
attributes = iris['data']
category = iris['target']
```

```python
clf = SVC()
```

```python
clf.fit(attributes, category)
```

```python
predict = clf.predict(attributes)
```

```python
fig1, (plt1, plt2) = plt.subplots(1, 2, sharey = True, figsize = (12, 4.5))

plt1.set_title("Predict Result", size = 15)
plt1.set_xlabel('sepal length')
plt1.set_ylabel('sepal width')
plt1.scatter(attributes[:,0], attributes[:,1], c = predict)

plt2.set_title("Predict Error", size = 15)
plt2.set_xlabel('sepal length')
plt2.set_ylabel('sepal width')
plt.scatter(attributes[:,0], attributes[:,1], c = predict-category)

fig1.suptitle('Sepal', size = 20, c = 'r')
fig1.show()
```

```python
fig2, (plt3, plt4) = plt.subplots(1, 2, sharey = True, figsize = (12, 4.5))

plt3.set_title("Predict Result", size = 15)
plt3.set_xlabel('petal length')
plt3.set_ylabel('petal width')
plt3.scatter(attributes[:,2], attributes[:,3], c = predict)

plt4.set_title("Predict Error", size = 15)
plt4.set_xlabel('petal length')
plt4.set_ylabel('petal width')
plt.scatter(attributes[:,2], attributes[:,3], c = predict-category)

fig2.suptitle('Petal', size = 20, c = 'r')
fig2.show()
```
